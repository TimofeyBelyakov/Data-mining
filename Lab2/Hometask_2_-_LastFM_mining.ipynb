```python
data_by_artists = data_countries.groupby(['user','sex','country'])['artist'].apply(lambda x: ','.join(x)).reset_index()
data_by_artists
```

```python
data_by_artists['country'].value_counts()
```

```python
data_by_artists['sex'].value_counts()
```

```python
dummy_data_by_artists = data_by_artists['artist'].str.get_dummies(',')
# dummy_data_by_sex = data_by_artists['sex'].str.get_dummies(',')
# dummy_data_by_country = data_by_artists['country'].str.get_dummies(',')

new_data_countries = pd.concat([data_by_artists.drop(['sex','country','artist'], axis=1), 
                                dummy_data_by_artists], axis = 1).set_index(['user'])
new_data_countries.head(10)
```

```python
new_data_countries.drop(['[unknown]'],axis=1,inplace=True)
```

```python
from mlxtend.frequent_patterns import apriori, association_rules, fpgrowth, fpmax
```

<h1>Apriori<h1>

```python
#При меньшем значении support слишком много данных (происходит ошибка)
frequent_itemsets = apriori(new_data_countries, min_support=0.03, use_colnames=True)
frequent_itemsets
```

```python
lift_threshold = association_rules(frequent_itemsets, metric="lift")['lift'].describe()['75%']
conf_threshold = association_rules(frequent_itemsets, metric="lift")['confidence'].describe()['75%']
leverage_threshold = association_rules(frequent_itemsets, metric="lift")['leverage'].describe()['75%']
conv_threshold = association_rules(frequent_itemsets, metric="lift")['conviction'].describe()['75%']
```

```python
rules_conf = association_rules(frequent_itemsets, metric="confidence", min_threshold=conf_threshold)
rules_conf
```

```python
rules_lift = association_rules(frequent_itemsets, metric="lift", min_threshold=lift_threshold)
rules_lift
```

```python
rules_lev = association_rules(frequent_itemsets, metric="leverage", min_threshold=leverage_threshold)
rules_lev
```

```python
rules_conv = association_rules(frequent_itemsets, metric="conviction", min_threshold=conv_threshold)
rules_conv
```

```python
rules_final = rules_conf[(rules_conf['lift'] > lift_threshold) &
                         (rules_conf['leverage'] > leverage_threshold) &
                         (rules_conf['conviction']>conv_threshold)]
rules_final
```
